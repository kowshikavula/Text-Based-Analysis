```python
import regex as re # Import regex library with alias re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text) # Define a function 'tokenize' to extract words from text
```

```python
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"  # Test the 'tokenize' function on a sample text
testtokens = tokenize(testtext)
print(testtokens)
```

```python
import nltk  # Import nltk library
nltk.download('stopwords')  # Download nltk stopwords
stopwords = set(nltk.corpus.stopwords.words('english'))  # Get a set of English stopwords
print(stopwords)
```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords] # Define a function 'remove_stop' to remove stopwords from a list of tokens
cleantokens= remove_stop(testtokens)  # Test the 'remove_stop' function on cleaned tokens
print(cleantokens)
```

```python
pipeline = [str.lower, tokenize, remove_stop]  # Define a pipeline of functions for text preparation
def prepare(text, pipeline):    # Define a function 'prepare' to apply the pipeline to a text
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
        return tokens
```

```python
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

```python
df['num_tokens'] = df['tokens'].map(len)  # Apply the 'prepare' function to the 'text' column and create new 'tokens' and 'num_tokens' columns
df.tail()   # Display the last 5 rows of the DataFrame
```

```python
from collections import Counter # Import Counter class from collections module
tokens = tokenize("She likes my cats and my cats like my sofa.")  # Test the Counter class on a sample list of tokens
testcount= Counter(tokens)
print(testcount)
```

```python
more_tokens = tokenize("She likes dogs and cats.") # Update the Counter with more tokens
testcount.update(more_tokens)
print(testcount)
```

```python
counter = Counter()  # Create a Counter for all tokens in the 'tokens' column
df['tokens'].map(counter.update)
print(counter.most_common(5)) # Display the 5 most common tokens
```

```python
# Create a DataFrame 'freq_df' from the Counter, filter tokens with frequency greater than or equal to 1
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq']) 
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)  # Display the top 10 tokens based on frequency
```

```python
freq_df.describe() # Display descriptive statistics for the 'freq' column in 'freq_df'
```

```python
# Plot a horizontal bar chart for the top 15 tokens
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
